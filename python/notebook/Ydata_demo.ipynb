### Large scale machine learning !

Ydata includes a version of the Downpour SGD algorithm.

```python
features = "hour,C1,banner_pos,site_id,site_domain,site_category,app_id,app_domain,app_category,device_id,device_ip,device_model,device_type,device_conn_type,C14,C15,C16,C17,C18,C19,C20,C21".split(",")
```

```python
X = train[features]
```

```python
y = train[["click"]]
```

```python
X
```

```python
y
```

#### Train !

Here the DSGDClassifier class follows the scikit-learn interface.

```python
from ydata.dsgd.dsgd import DSGDClassifier
```

```python
clf = DSGDClassifier(hsc)
```

```python
clf.fit(X, y)
```

```python
ypred = clf.predict(X)
```

```python
ypred
```

#### Test !

Predictions on test sample

```python
X_test = test[features]
```

```python
id_test = test[["id"]]
```

```python
X_test
```

```python
ypred_test = clf.predict(X_test)
```
